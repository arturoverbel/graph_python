```python
def probability(x, p):
    x = np.array(x)
    p = np.array(p)
    y = np.random.rand()

    cdf = np.cumsum(p)
    total = len(x)

    pos = 0

    for n in range(total - 1):
        pos = pos + (y >= cdf.item(n))

    return x.item(pos)

result = probability([0, 1, 3, 4], [0.1, 0.2, 0.1, 0.6])

print(result)

```
